# Change Values to 1/0
# Take all values of 'overturned' and 'pardoned'

If values is True change to 1

If values is False change to 0


```python
# overturned code
code_overturned_for_True= "True"
code_overturned_for_False = "False"
# pardoned code
code_pardoned_for_True= "True"
code_pardoned_for_False = "False"
# binary value
value_of_one = 1
value_of_zero = 0

def change_code_overturned_to_biner(overturned):
    return value_of_one if overturned == code_overturned_for_True else value_of_zero

def change_code_pardoned_to_biner(pardoned):
    return value_of_one if pardoned == code_pardoned_for_True else value_of_zero
```

```python
# Update all values of 'overturned' series
data["overturned"] = data["overturned"].apply(change_code_overturned_to_biner)

# Update all values of 'pardoned' series
data["pardoned"] = data["pardoned"].apply(change_code_pardoned_to_biner)
```
